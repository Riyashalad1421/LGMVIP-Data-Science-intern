# Finding the best K-value

```python
l_test=[]
l_train=[]

for i in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    x=confusion_matrix(y_test,knn.predict(X_test))
    y=confusion_matrix(y_train,knn.predict(X_train))
    l_test.append((x[0][1]+x[0][2]+x[1][0]+x[1][2]+x[2][0]+x[2][1])/x.sum())
    l_train.append((y[0][1]+y[0][2]+y[1][0]+y[1][2]+y[2][0]+y[2][1])/y.sum())
```

```python
print(l_test)

print(l_train)
```

```python
plt.plot(range(1,30),l_test,label="test-error")
plt.plot(range(1,30),l_train,label="train-error")
plt.xlabel("K-values")
plt.ylabel("Error")
plt.legend()
```

# cross_validation

```python
x_cvtrain,x_cvtest,y_cvtrain,y_cvtest=train_test_split(X_train,y_train,test_size=0.3)
```

```python
l_cvtest=[]
l_cvtrain=[]

for i in range(1,20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_cvtrain,y_cvtrain)
    xc=confusion_matrix(y_cvtest,knn.predict(x_cvtest))
    yc=confusion_matrix(y_cvtrain,knn.predict(x_cvtrain))
    l_cvtest.append((xc[0][1]+xc[0][2]+xc[1][0]+xc[1][2]+xc[2][0]+xc[2][1])/xc.sum())
    l_cvtrain.append((yc[0][1]+yc[0][2]+yc[1][0]+yc[1][2]+yc[2][0]+yc[2][1])/yc.sum())

    
print(l_cvtrain)
```

```python
plt.plot(range(1,20),l_cvtest,label="cvtesterror")
plt.plot(range(1,20),l_cvtrain,label="cvtrainerror")
plt.legend()
```

```python
knncv=KNeighborsClassifier(n_neighbors=3)
knncv.fit(x_cvtrain,y_cvtrain)
```
